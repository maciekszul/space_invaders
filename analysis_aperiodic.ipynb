```python
from scipy.fftpack import fft, fftfreq, fftshift
import matplotlib.pylab as plt
import matplotlib.patches as mpatches
import numpy as np
import pandas as pd
from scipy.interpolate import interp1d
# Import spectral power functions
from neurodsp.spectral import compute_spectrum, rotate_powerlaw

# Import utilities for loading and plotting data
from neurodsp.utils import create_times
from neurodsp.utils.download import load_ndsp_data
from neurodsp.plts.spectral import plot_power_spectra
from neurodsp.plts.time_series import plot_time_series
```

```python
def resamp_interp(x, y, new_x):
    """
    returns resampled an interpolated data
    """
    resamp = interp1d(x, y, kind='linear', fill_value='extrapolate')
    new_data = resamp(new_x)
    return new_data
```

```python
data = pd.read_csv("/home/mszul/git/space_invaders/data/000_1620029633.856371_traj.csv")
line = np.load("/home/mszul/git/space_invaders/data/000_1620029633.856371_line.csv.npy")
f, (ax1, ax2) = plt.subplots(2, 1, figsize=(10,5), dpi=200, sharex="all")
zigzag_pos = -data.zigzag_pos - 5
line_new = resamp_interp(line[:,1], line[:,0], zigzag_pos)
time = data.time - data.time[0]
time = time.to_numpy()
ax1.axhline(0, c="black", lw=0.3)
ax1.plot(time, data.cursor_pos - line_new, lw=0.3)
ax1.plot(time, line_new, lw=0.3)
ax1.plot(time, data.cursor_pos, lw=0.3)
ax1.set_ylim(-10, 10)
x = data.cursor_pos - line_new
x = x.to_numpy()
width = 120
hhh = []
for i in range(len(x)):
    try:
        h = np.pad(x[0:i], [width-i, 0])
    except:
        h = x[i-width:i]
    fft_Y = fft(h, axis=0)
    fft_FREQ = fftfreq(fft_Y.shape[0], 1/60)
    fft_Y = fftshift(fft_Y)
    fft_FREQ = fftshift(fft_FREQ)
    freqs = np.where(fft_FREQ >= 0)[0]
    hhh.append(np.abs(np.real(fft_Y))[freqs])
ax2.imshow(np.vstack(hhh).transpose()[:6], aspect="auto", cmap="Blues", extent=[time[0], time[-1], 5, 0])
ax2.set_xlabel("time [s]")
ax2.set_ylabel("freq [Hz]")
ax1.set_ylabel("distance[deg va]")
```
